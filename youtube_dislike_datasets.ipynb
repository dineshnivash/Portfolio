```python
# 5) Convert datatype of column published_at from object to pandas datetime.
import datetime as dt #importing datetime funtion to change the object into datetime format
you_tube['published_at']=you_tube['published_at'].apply(lambda x: dt.datetime.strptime(x,'%d-%m-%Y %M:%S'))
```

```python
# 6) Create a new column as 'published_month' using the column published_at (display the months only)
you_tube['published_month']=you_tube['published_at'].dt.month #using dt.month to seperate a month in the published_at column
```

```python
# 7) Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....
month_name={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
you_tube['published_month']=you_tube['published_month'].replace(month_name) #using dictionary and assigning 1 as jan, 2 as feb..., then replacing that as jan,feb...
```

```python
# 11) Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.
you_tube.groupby(by='title')['likes'].max().sort_values(ascending=False).head(1)
#by using groupby method found that BTS() 'Dynamite' Official MV  videos has the maximum number of likes (31837675)
#in this using a groupby method we are grouping the title and getting the likes of max number using max(), then sort the values using sort_values of selected group in descending order and getting top one title by head(1) 
```

```python
you_tube.groupby(by='title')['likes'].min().sort_values(ascending=False).tail(1)
#by using groupby method found that Leading the Charge | Circle K  videos has the minimum number of likes (22)
#in this using a groupby method we are grouping the title and getting the likes of min number using min(), then sort the values using sort_values of selected group in descending order and getting lowest one title by tail(1)
```

```python
# 12) Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes and write your inferences.
you_tube.groupby(by='title')['dislikes'].max().sort_values(ascending=False).head(1)
#by using groupby method found that Cuties | Official Trailer | Netflix videos has the maximum number of dislikes (2397733)
#in this using a groupby method we are grouping the title and getting the dislikes of max number using max(), then sort the values using sort_values of selected group in descending order and getting top one title by head(1) 
```

```python
you_tube.groupby(by='title')['dislikes'].min().sort_values(ascending=False).tail(1)
#by using groupby method found that Tims For Good: A Taste Of The Familiar videos has the minimum number of dislikes (3)
#in this using a groupby method we are grouping the title and getting the dislikes of min number using min(), then sort the values using sort_values of selected group in descending order and getting lowest one title by tail(1)
```
